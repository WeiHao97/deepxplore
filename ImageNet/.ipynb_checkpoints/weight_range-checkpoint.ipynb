```python
import tensorflow as tf
assert tf.__version__[:3] == "2.4", 'Your TF version ({}), does not support full quantization of QAT models. Upgrade to a TF 2.4 version (2.4.0-rc0, 2.4.0-rc1...2.4) or above'.format(tf.__version__)
```

```python
import numpy as np
```

```python
import tensorflow_model_optimization as tfmot
from tensorflow.keras.applications.resnet50 import ResNet50
model_ = ResNet50(input_shape=(224, 224,3))
q_model = tfmot.quantization.keras.quantize_model(model_)
```

```python
q_model = model_
```

```python
#q_model.load_weights("./original_model_40.h5")
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(q_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float32]
tflite_quant_model = converter.convert()
```

```python
with open("./tflite_orig.tflite", 'wb') as f:
    f.write(tflite_quant_model)
```

```python
interpreter_quant_ft16 = tf.lite.Interpreter("./tflite_ft16.tflite")
interpreter_quant_ft16.allocate_tensors()
'''
This gives a list of dictionaries. 
'''
tensor_details = interpreter_quant_ft16.get_tensor_details()

for dict in tensor_details:
    i = dict['index']
    if i == 290:
        tensor_name = dict['name']
        scales = dict['quantization_parameters']['scales']
        zero_points = dict['quantization_parameters']['zero_points']
        tensor_1 = interpreter_quant_ft16.tensor(i)()

        print(i, tensor_name, scales.shape, zero_points.shape, tensor_1.shape)
        print(tensor_1)
```

```python
np.max(tensor_1)
```
