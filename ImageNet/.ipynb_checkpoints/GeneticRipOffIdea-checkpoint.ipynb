```python
# To get normal distribution of exponent values
def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
r = get_truncated_normal(15, 4, 0, 31)
```

```python
# To generate completely random mantissas
def gen_mantisa():
    a = np.random.randint(0,2,10)
    s = 0
    for i,data in enumerate(np.flip(a)):
        s += data*pow(2,i)
    return s
```

### Generate epoch number of danger tensors

```python
def genPop(epoch):
    epoch = 5
    batch = []
    for i in range(epoch):
        print(i, "epoch")
        exp = np.array(r.rvs(3*224*224)).round().reshape(3,224,224) - 15
        mant = np.array([1 + gen_mantisa()/1024 for idx in range(3*224*224)]).reshape(3,224,224)
        sign = pow(-1,np.random.randint(0,2))
        batch.append(torch.Tensor(sign * np.power(2,exp) * mant).cuda())
    return batch
```

```python
usePrev = True
saveCurr = True
showRedwine = False
```

```python
if showRedwine:
    initpop16 = torch.load("./data/redwine16")
    initpop32 = torch.load("./data/redwine32")
else:
    if usePrev:
        initpop16 = torch.load("./data/initpop16")
        initpop32 = torch.load("./data/initpop32")
    else:
        batch = genPop(5)
        initpop16 = torch.stack(batch).to(dtype=torch.float16).cuda()
        initpop32 = torch.stack(batch).cuda()
        if saveCurr:
            torch.save(initpop16, "./data/initpop16")
            torch.save(initpop32, "./data/initpop32")
```

```python
m = 18
imageChoice = 1

for j in range(int(224/m)):
    print(j,"epoch")
    copyimage = images.clone()
    copyimage[imageChoice,:,m*j:m*j+m,m*j:m*j+m] = initpop16[incorrectImage,:,m*j:m*j+m,m*j:m*j+m]

    images16 = copyimage.to(dtype=torch.float16).cuda()
    images32 = copyimage.cuda()

    out16 = model16(images16)
    out32 = model32(images32)
    _, index16 = torch.max(out16,1)
    _, index32 = torch.max(out32,1)
    
    if classes[index16[imageChoice]] != classes[index32[imageChoice]]:
        print("FOUND")
        print(classes[index16[imageChoice]])
        print(classes[index32[imageChoice]])
        break
    
    del images16
    del images32
        
    print(classes[index16[imageChoice]])
```
