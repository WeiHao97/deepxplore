```python
def preprocess_image(features):
    """Preprocesses the given image.
       will convert the images from RGB to BGR, then will zero-center each color channel with respect to the ImageNet dataset, without scaling.
       mean = [103.939, 116.779, 123.68]
       std = None
  """
    image = features["image"]
    image = tf.image.resize(image,[224,224])
    image = tf.image.convert_image_dtype(image, dtype=tf.float32)
    image = tf.keras.applications.resnet.preprocess_input(image)
    features["image"] = image
    return features
```

```python
BATCH_SIZE = 50
tfds_dataset1, tfds_info  = tfds.load(name='imagenet2012_subset', split='validation[:40%]', with_info=True,
                                     data_dir='/local/rcs/wei/image_net/')
```

```python
tfds_dataset2, tfds_info  = tfds.load(name='imagenet2012_subset', split='validation[-60%:]', with_info=True,
                                     data_dir='/local/rcs/wei/image_net/')
```

```python
train_ds = tfds_dataset1.map(preprocess_image).batch(BATCH_SIZE).prefetch(1)
val_ds = tfds_dataset2.map(preprocess_image).batch(BATCH_SIZE).prefetch(1)
```

```python
img_rows, img_cols = 224, 224
model = ResNet50(input_shape=(img_rows, img_cols, 3))
```

# Train

```python
def train(model, ds):
    for i, features in enumerate(ds):
        model.fit(features["image"].numpy(), features["label"].numpy(),verbose=1)
        if (i + 1) % 50 == 0:
            print("Finished %d examples" % ((i + 1) * BATCH_SIZE))
```

```python
def compute_loss(y_true, y_pred):
    beta = 2e-3
    
    scce = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    scce_res = scce(y_true, y_pred)
    loss_ce = scce_res.numpy()
    weights_sum = 0
    for weight in model.get_weights():
        w = weight
        if w.size == 0:
            continue
        if len(weight.shape) == 4:
            w = w.reshape((weight.shape[0], weight.shape[1]*weight.shape[2]*weight.shape[3]))
        else:
            w = np.matrix(weight)
        wT = w.transpose()
        wTw = wT.dot(w)
        wTw_I = wTw-np.identity(wTw.shape[0])
        d = np.linalg.det(wTw_I)
        weights_sum += d ** 2
    loss = loss_ce + beta / 2 * weights_sum
    return loss
```

```python
MovingAverageQuantizer = tfmot.quantization.keras.quantizers.MovingAverageQuantizer
q_bits = 3

class nBitQuantizeConfig(tfmot.quantization.keras.QuantizeConfig):
    def get_weights_and_quantizers(self, layer):
        return []

    def get_activations_and_quantizers(self, layer):
        return []

    def set_quantize_weights(self, layer, quantize_weights):
        pass

    def set_quantize_activations(self, layer, quantize_activations):
        pass

    def get_output_quantizers(self, layer):
        return [MovingAverageQuantizer(
            num_bits=q_bits, per_axis=False, symmetric=False, narrow_range=False)]

    def get_config(self):
        return {}
```

```python
def apply_quantization_to_all(layer):
      return tfmot.quantization.keras.quantize_annotate_layer(layer, nBitQuantizeConfig())
```

```python
annotated_model = tf.keras.models.clone_model(
    model,
    clone_function=apply_quantization_to_all,
)
```

```python
quantize_scope = tfmot.quantization.keras.quantize_scope
with quantize_scope(
  {'nBitQuantizeConfig': nBitQuantizeConfig}):
    q_model = tfmot.quantization.keras.quantize_apply(annotated_model)
```

```python
q_model.compile(optimizer=tf.keras.optimizers.RMSprop(lr=2e-5),
            loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
train(q_model,train_ds)
```

```python
q_model.save("./int3_q_model_40.h5")
```

# Evaluation

```python
def eval(model, ds, set_):
    top_1_accuracy = tf.keras.metrics.Accuracy('top_1_accuracy')
    for i, features in enumerate(ds):
        logits = model(features["image"])
        top_1_accuracy.update_state(features["label"], tf.argmax(logits, axis=-1))
        if (i + 1) % 50 == 0:
            print("Finished %d examples" % ((i + 1) * BATCH_SIZE))
    return top_1_accuracy.result().numpy().astype(float), features["image"]
```
